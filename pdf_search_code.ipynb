```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

```

```python
# 1. loadingPDF
loader = PyPDFLoader("dosya.pdf")
documents = loader.load()

print(f"âœ… PDF yÃ¼klendi, toplam {len(documents)} sayfa bulundu.")
```

```python
# 2. Metin parÃ§alama
text_splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=50)
docs = text_splitter.split_documents(documents)

print(f"âœ… Metin {len(docs)} parÃ§aya ayrÄ±ldÄ±.")

# 3. Embedding model
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

# 4. FAISS index oluÅŸturma
db = FAISS.from_documents(docs, embeddings)
print("âœ… FAISS index hazÄ±r.")
```

```python
#Search
query = "machine learning"  # Ã¶rnek sorgu
results = db.similarity_search(query, k=5)

print("\nðŸ”Ž Semantic Search SonuÃ§larÄ±:")
for r in results:
    print("-", r.page_content[:200], "...")  #ilk 200 karakter
```
